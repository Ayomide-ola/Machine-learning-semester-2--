### 1.1 What is data
A data is a compilation of plain facts that have been gathered within a defined context. Statisticians would refer to it as a set of observations contained in variables or columns of varying/unique characteristics. As earlier defined that machien learning broadly as a concept is a brabch of artificial intelligence that involes using the information we have today(Data) to make predictions about tommorrow. ***a. Classification:*** Classification supervised ML is the type of model that predicts categorical targets. This type of model focuses on predicting it the furture will be a certain class or not e.g 'will it rain tomorrow?', model then classfy 'Yes OR No'. 
***b. Regression:*** In regression supervised ML, the model predicts a contineous targets. This type of model focuses on predicting it the future will be certain numerics or other. For example: 'What will the house price be tommorw', 'Model: $400,000' (Banoula, 2023).

2. Unsupervised machine learning: As the name implies, 'Unsupevised' learning in simple terms means the algorithm learns under no supervision. This means the machien is not explicitly given a target. The models derive their own target from the variance in the data. In Unsupervised Learning, the machine uses unlabeled data and learns on itself without any supervision. The machine tries to find a pattern in the unlabeled data and gives a response (Banoula, 2023).

### Measure of central tendency (Location)
This gives an insight into the distribution of data. The central tendency shows the middle of the data, and to analyse this we use the 5 number summary. 
##### Five Number summary
Five number summary means summarising data under five heads namely: Mean, median, Mode, Min and Max. 
1. ***Mean:*** This is the average of the observations in a column. It is calculated by summing together all observations in a variable and dividing the sum by the count of all observation. In 'seed' the mean of any variable would be the sum of all variables divided by 320. It is the most common indicator of central tendency (Torres-Reyna, 2008)
2. ***Median:*** This is the middle or average of the middle observations of a variable after the observations have been sorted in acessending order, i.e from the lowest observation to the highest. This in our number summary would be the 50% of the variable. It is another measure of central tendency (Torres-Reyna, 2008).
3. ***Mode:*** Mode refers to the most frequent observation in a given variable. 
4. ***Min:*** This is, simply, the minimum value in each variable. 
5. ***Max:*** This is the maximum value in each variable. 
With this five statistical operations, we can summarise the locations of observations in each variable.
#### Variability
In addition to knowing the locations of our observations, it is important to explore the variability of the observations from the mean (50%). To do this we would have to calculate: Variance, range, and Standard deviation. 

6. ***Variance:*** Variance, as the name implies, is the measure of dispersion of data from the mean (50%) (Torres-Reyna, 2008). This is calculated by squaring the difference between mean and each observation in a column i.e (each column observation minus mean of the column), and dividing the summation of squares by the total number of observation minus 1 i.e (320-1) for 'seed' (Hayes, 2023). 
7. ***Standard deviation:*** Standard deviation is the square root of variance. Variables with high standard deviation relative to the mean have their data spread out from the mean, while those with small standard deviation have their data tightly grouped close to the mean.

#### Locating outliers

```python
from matplotlib.figure import Figure
from matplotlib.backends.backend_agg import FigureCanvas 
%matplotlib inline
```

```python
def mpl_plot(avg, highlight):# Creating a function that works for all
    fig = Figure() # Create a visualisation
    FigureCanvas(fig) # Create a canvas 
    ax = fig.add_subplot() # Create a subplot 
    avg.plot(ax=ax) # This plots line plot, scatter for example will be avg. scatter
    if len(highlight): highlight.plot(style='o', ax=ax)  #Plot if I have outliers, and if there are no outliers don't plot
    return fig 
def find_outliers(variable = 'Grain-yeild', window = 30, sigma = 10, view_fn = mpl_plot):# variable, the time window for exponential, sigma= 0.3(normally): sigma value is what controls the outer range 
    avg = seed[variable].rolling(window =window).mean()# The bigger the time window the smother the graph
    residual = seed[variable]-avg                       # When I increase the value of sigma it makes outliers more, what plot
    std = residual.rolling(window = window).std()
    outliers = (np.abs(residual)>std*sigma)
    return view_fn(avg, avg[outliers])
```

#### Implementing Shillouttes score metrics

### Shilloutte score:
The Silhouette Coefficient is calculated using the mean intra-cluster distance (a) and the mean nearest-cluster distance (b) for each sample. The Silhouette Coefficient for a sample is (b - a) / max(a, b). To clarify, b is the distance between a sample and the nearest cluster that the sample is not a part of. Note that Silhouette Coefficient is only defined if number of labels is 2 <= n_labels <= n_samples - 1.

This function returns the mean Silhouette Coefficient over all samples. To obtain the values for each sample, use silhouette_samples.

The best value is 1 and the worst value is -1. Values near 0 indicate overlapping clusters. Negative values generally indicate that a sample has been assigned to the wrong cluster, as a different cluster is more sim  (SKlearn, 2024).lar.

### Testing the Stationarity of the Dataset
Obviously the plot shows that the pattern in the observation are not stationary. However in other to assertain the stationarity of the dataset using Augmented Dickey-Fuller Test. Statistical test that can help determine if the time series is stationary. The ADF test is a type of unit root test. Unit roots are a cause for non-stationarity, the ADF test will test if the unit root is present. 

#### Defining the Hypothesis:
1. Null Hypothesis(H0):Time sreries is not stationary.
2. Altanate Hypothesis (Ha): Time series is stationary. 

If the p-value is less than the Significance Level defined, we reject the Null Hypothesis that the time series contains a unit root. In other words, by rejecting the Null hypothesis, we can conclude that the time series is stationary.

```python
from statsmodels.tsa.stattools import adfuller
```

```python
# Augmented Dickey-Fuller test (ADF Test)
ad_fuller_result = adfuller(Btc['Close'])

# Display the outcome of Augmented Dicky-Fuller Test
print(f'ADF Statistic: {ad_fuller_result[0]}')
print(f'p-value: {ad_fuller_result[1]}')
```
